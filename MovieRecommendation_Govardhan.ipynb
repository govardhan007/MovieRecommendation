```python
import pandas as pd
import collections
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
matrix = np.zeros((max(testgetmax[1], traingetmax[1])+1,max(testgetmax[0], traingetmax[0])+1 ))
for i in range(0, df2tot):
  userID, itemID, rating, name = df2.iloc[i]
  matrix[itemID][userID] = rating
cosmatrix = np.zeros((max(testgetmax[1], traingetmax[1])+1,max(testgetmax[0], traingetmax[0])+1))
k = max(testgetmax[1], traingetmax[1])+1
l = max(testgetmax[0], traingetmax[0])+1
testvalue = 5000
for i in range(1, k):
  sum = np.sum(matrix[i])
  non_zero = np.count_nonzero(matrix[i])
  if sum == 0.0:
    continue
  avg = sum/non_zero
  for j in range(1, l):
    if matrix[i][j] != 0:
      cosmatrix[i][j] = matrix[i][j]
```

```python
simmatcosine = np.zeros((max(testgetmax[1], traingetmax[1])+1,max(testgetmax[1], traingetmax[1])+1))
```

```python
#calculating similarity of values
import math
from math import sqrt
def SimVal(a, b) -> float:
  dot = np.dot(a,b)
  first = (math.sqrt(np.sum(np.square(a))))
  second = (math.sqrt(np.sum(np.square(b))))
  mult = first * second
  if mult == 0.0:
    return 0.0
  val = dot/(first*second)
  return val
for i in range(1,k):
  for j in range(1,k):
    simmatcosine[i][j] = SimVal(cosmatrix[i], cosmatrix[j])
def RMSE(x, y):
  return round(sqrt(mean_squared_error(x, y)),4)
```

Cosine Similarity


```python
#predicting ratings for the item_id's
predictedratings = list()
orgratings = list()
for i in range(0, testvalue):
  predict_user, predict_movie, actual_rating, movie_name = df1.iloc[i]
  orgratings.append(actual_rating)
  cosdict = collections.defaultdict(float)

  for i in range(1,k):
    if i != predict_movie and simmatcosine[predict_movie][i] > 0.2:
      cosdict[i] = simmatcosine[predict_movie][i]

  cosine_value_sum = 0.0
  cosine_dot_sum = 0.0
  for key, value in cosdict.items():
    if matrix[key][predict_user] == 0.0:
      continue
    cosine_value_sum = cosine_value_sum + value
    cosine_dot_sum = cosine_dot_sum + matrix[key][predict_user]*value

  if cosine_value_sum == 0.0:
    predicted_cosine_value = 0.0
  else:
    predicted_cosine_value = cosine_dot_sum/cosine_value_sum

  predictedratings.append(predicted_cosine_value)
```

```python
#comparing actual Vs Predicted rating values
rmslist = list()
predictedrmslist = list()

for i in range(0, testvalue):
  if predictedratings[i] > 0.0:
    rmslist.append(orgratings[i])
    predictedrmslist.append(predictedratings[i])

cosine_lst_tuple = list(zip(rmslist, predictedrmslist))

print("(actual,predicted")
print(cosine_lst_tuple)
```

```python
#rmse value obtained
print(RMSE(rmslist, predictedrmslist))
```
